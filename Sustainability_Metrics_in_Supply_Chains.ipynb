###1. Delivery Time and Late Delivery Rate

Here we measure the average dats for shipping in real vs the scheduled data available to us in order to check the efficiency of the supply chain. We will calculate the late delivery rate with the data provided and it will help indicate better sustainability in terms of customer satisfaction and resource utlization.

```python

# Calculate the average Days for Shipping (real) and Days for Shipment (scheduled)
average_real_shipping_days = df_scd['Days for shipping (real)'].mean()
average_scheduled_shipping_days = df_scd['Days for shipment (scheduled)'].mean()

# Compute the late delivery rate
late_delivery_rate = df_scd['Late_delivery_risk'].mean()

# Display the results
print(f"Average Days for Shipping (real): {average_real_shipping_days:.2f} days")
print(f"Average Days for Shipment (scheduled): {average_scheduled_shipping_days:.2f} days")
print(f"Late Delivery Rate: {late_delivery_rate:.2%}")

# Create a bar plot to visualize the average shipping days
plt.figure(figsize=(10, 6))
sns.barplot(x=['Real Shipping Days', 'Scheduled Shipping Days'],
            y=[average_real_shipping_days, average_scheduled_shipping_days])
plt.title('Average Shipping Days')
plt.ylabel('Days')
plt.show()

# Create a pie chart to visualize the late delivery rate
plt.figure(figsize=(8, 8))
labels = ['On Time', 'Late']
sizes = [1 - late_delivery_rate, late_delivery_rate]
colors = ['#ff9999', '#66b3ff']
explode = (0.1, 0)  # explode the 1st slice (i.e., 'Late')
plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%', shadow=True, startangle=90)
plt.title('Late Delivery Rate')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

```python


# Group the data by 'Order Region' and calculate average delivery times
average_delivery_times = df_scd.groupby('Order Region')[['Days for shipping (real)', 'Days for shipment (scheduled)']].mean()

# Group the data by 'Order Region' and 'Category Name' to calculate late delivery rate
late_delivery_rates = df_scd.groupby(['Order Region', 'Category Name'])['Late_delivery_risk'].mean().reset_index()

# Display the average delivery times
print("Average Delivery Times by Order Region:")
print(average_delivery_times)

# Display the late delivery rates
print("\nLate Delivery Rates by Order Region and Category Name:")
print(late_delivery_rates)

# Create a bar plot to visualize the average delivery times by region
plt.figure(figsize=(12, 6))
sns.barplot(data=average_delivery_times, x=average_delivery_times.index, y='Days for shipping (real)', palette='viridis')
plt.title('Average Real Shipping Days by Order Region')
plt.xticks(rotation=90)
plt.ylabel('Days')
plt.show()

# Create a bar plot to visualize the late delivery rates by region and category
plt.figure(figsize=(12, 6))
sns.barplot(data=late_delivery_rates, x='Order Region', y='Late_delivery_risk', hue='Category Name', palette='Set1')
plt.title('Late Delivery Rates by Order Region and Category Name')
plt.xticks(rotation=90)
plt.ylabel('Late Delivery Rate')
plt.legend(title='Category Name', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

###2. Profitability and Efficiency

Here we analyse the benefit per order and order profit per order through which we will judge the economic sustainability of the supply chain. We know that higher profits for each order will indicate better cost management and resource utilization. We will also understand the profitability of individual items in supply chain.

```python


# Calculate Benefit per Order (BPO) and Order Profit Per Order (OPO)
df_scd['Benefit per Order'] = df_scd['Sales'] - df_scd['Order Item Total']
df_scd['Order Profit Per Order'] = df_scd['Order Profit Per Order']

# Calculate Order Item Profit Ratio (OIPR)
df_scd['Order Item Profit Ratio'] = df_scd['Order Item Profit Ratio']

# Find the highest profitable item region-wise
highest_profitable_items = df_scd.groupby(['Order Region', 'Product Name'])[['Benefit per Order', 'Order Profit Per Order']].sum().reset_index()
highest_profitable_items = highest_profitable_items.sort_values(by=['Order Region', 'Benefit per Order'], ascending=[True, False])
highest_profitable_items = highest_profitable_items.drop_duplicates(subset='Order Region')

# Display the highest profitable items region-wise
print("Highest Profitable Items Region-wise:")
print(highest_profitable_items)

# Create a bar plot to visualize Benefit per Order (BPO) by region with product names
plt.figure(figsize=(12, 6))
sns.barplot(data=df_scd, x='Order Region', y='Benefit per Order', hue='Product Name', palette='viridis', ci=None)
plt.title('Benefit per Order (BPO) by Order Region')
plt.xticks(rotation=90)
plt.ylabel('Benefit per Order')
plt.legend(title='Product Name', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()

# Create a bar plot to visualize Order Profit Per Order (OPO) by region with product names
plt.figure(figsize=(12, 6))
sns.barplot(data=df_scd, x='Order Region', y='Order Profit Per Order', hue='Product Name', palette='viridis', ci=None)
plt.title('Order Profit Per Order (OPO) by Order Region')
plt.xticks(rotation=90)
plt.ylabel('Order Profit Per Order')
plt.legend(title='Product Name', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()

# Create a bar plot to visualize Order Item Profit Ratio (OIPR) by region with product names
plt.figure(figsize=(12, 6))
sns.barplot(data=df_scd, x='Order Region', y='Order Item Profit Ratio', hue='Product Name', palette='viridis', ci=None)
plt.title('Order Item Profit Ratio (OIPR) by Order Region')
plt.xticks(rotation=90)
plt.ylabel('Order Item Profit Ratio')
plt.legend(title='Product Name', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

###3. Resource Utilization

Here we will calcualte which customer segment generated the most sales. Marketing efforts and resource wastage will be reduced significantly due to the information generated here. We will also study the order item quantity to assess product demand and inventory management


```python

# Calculate Sales per Customer
sales_per_customer = df_scd.groupby('Customer Id')['Sales'].sum()

# Calculate Sales per Customer Segment
sales_per_segment = df_scd.groupby('Customer Segment')['Sales'].sum()

# Display the Sales per Customer and Sales per Customer Segment
print("Sales per Customer:")
print(sales_per_customer.sort_values(ascending=False).head(10))  # Display the top 10 customers by sales
print("\nSales per Customer Segment:")
print(sales_per_segment)

# Create a bar plot to visualize Sales per Customer Segment
plt.figure(figsize=(10, 6))
sns.barplot(data=df_scd, x='Customer Segment', y='Sales', estimator=sum, ci=None, palette='viridis')
plt.title('Sales per Customer Segment')
plt.xticks(rotation=45)
plt.ylabel('Total Sales')
plt.show()

# Analyze Order Item Quantity with respect to Product Name
product_demand = df_scd.groupby('Product Name')['Order Item Quantity'].sum().sort_values(ascending=False).head(10)

# Display the top 10 products by demand
print("\nTop 10 Products by Demand:")
print(product_demand)

# Create a bar plot to visualize Order Item Quantity for the top 10 products by demand
plt.figure(figsize=(12, 6))
sns.barplot(x=product_demand.index, y=product_demand, palette='viridis')
plt.title('Top 10 Products by Demand')
plt.xticks(rotation=90)
plt.ylabel('Order Item Quantity')
plt.xlabel('Product Name')
plt.show()
```

###4. Shipping Modes
Here we will evaluate the distribution of orders across different Shipping Modes (e.g., Standard Class, First Class) to identify opportunities for more sustainable shipping options, such as promoting eco-friendly modes etc.


```python

# Count the number of orders for each shipping mode
shipping_mode_counts = df_scd['Shipping Mode'].value_counts()

# Display the distribution of orders across shipping modes
print("Distribution of Orders by Shipping Mode:")
print(shipping_mode_counts)

# Create a bar plot to visualize the distribution of orders by shipping mode
plt.figure(figsize=(8, 6))
sns.barplot(x=shipping_mode_counts.index, y=shipping_mode_counts.values, palette='viridis')
plt.title('Distribution of Orders by Shipping Mode')
plt.xlabel('Shipping Mode')
plt.ylabel('Number of Orders')
plt.xticks(rotation=45)
plt.show()
```

### 5. Geographical Analysis
Here we will analyze the markets and regions where orders are delivered to identify areas with high demand. This information can help optimize distribution and reduce unnecessary transportation emissions.

```python
# Group the data by Market and Order Region to calculate total sales
market_region_sales = df_scd.groupby(['Market', 'Order Region'])['Sales'].sum().reset_index()

# Pivot the data for better visualization
pivot_market_region_sales = market_region_sales.pivot(index='Order Region', columns='Market', values='Sales')

# Create a heatmap to visualize sales by Market and Order Region
plt.figure(figsize=(12, 8))
sns.heatmap(pivot_market_region_sales, cmap='YlGnBu', annot=True, fmt=".0f", linewidths=.5)
plt.title('Sales by Market and Order Region')
plt.xlabel('Market')
plt.ylabel('Order Region')
plt.xticks(rotation=45)
plt.show()
```

### 6. Order Status Analysis
Here we will measure the distribution of Order Status (e.g., COMPLETE, CANCELED) to identify trends related to order fulfillment and customer satisfaction.

```python
# Calculate the distribution of Order Status
order_status_counts = df_scd['Order Status'].value_counts()

# Create a bar plot to visualize the distribution of Order Status
plt.figure(figsize=(10, 6))
sns.barplot(x=order_status_counts.index, y=order_status_counts.values, palette='viridis')
plt.title('Distribution of Order Status')
plt.xticks(rotation=45)
plt.ylabel('Number of Orders')
plt.xlabel('Order Status')
plt.show()

# Calculate the percentage distribution of Order Status
order_status_percentages = order_status_counts / order_status_counts.sum() * 100

# Display the percentage distribution of Order Status
print("Percentage Distribution of Order Status:")
print(order_status_percentages)
```
